```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator

# Create a Quantum Circuit with 2 qubits and 2 classical bits
qc = QuantumCircuit(2, 2)

# Apply a Hadamard gate to qubit 0 to put it into superposition
qc.h(0)

# Apply a CNOT gate with control qubit 0 and target qubit 1 to entangle them
qc.cx(0, 1)

# Measure both qubits into their respective classical bits
qc.measure([0, 1], [0, 1])

# Simulate the circuit locally
simulator = AerSimulator()
result = simulator.run(qc, shots=1024).result()
counts = result.get_counts()

print("Circuit Design:")
print(qc.draw(output='text'))
print(counts)
```
